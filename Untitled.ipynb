# IQR


```python
q1=df['goals_z'].quantile(0.25)
q3=df['goals_z'].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
IQR
```

```python
lower_range=q1-1.5*IQR
```

```python
lower_range
```

```python
upper_range = q3+1.5*IQR
```

```python
upper_range
```

Trimming

```python
Td=df[(df['goals_z']>lower_range)&(df['goals_z']>upper_range)]
```
